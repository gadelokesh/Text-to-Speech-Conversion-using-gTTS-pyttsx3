# Text to Speech conversion using GTTS

```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS('''Hi I am L and Welcome to my Github , Here You can Explore lot of Projects and Resources for learning lot of things, Hope you like it,Thank you''',lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file, autoplay= False)
```

# Text to speech conversion using pyttsx3

## We can change Gender using pyttsx3

```python
import pyttsx3
from IPython.display import Audio

text = '''Hi I am L and Welcome to my Github , Here You can Explore lot of Projects and Resources for learning lot of things, Hope you like it,Thank you.'''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

# change the voices
voice = audio.getProperty('voices')

# 0 for male and 1 for female
audio.setProperty('voice', voice[0].id)
#audio.setProperty('voice', voice[1].id)

# text-to speech conversion
audio.say(text)

# save the audio file
audio.save_to_file(text, 'test_male_Voice.mp3')
#audio.save_to_file(text, 'test_female_Voice.mp3')

audio.runAndWait()
```
